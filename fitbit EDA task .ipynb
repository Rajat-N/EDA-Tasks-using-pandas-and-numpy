```python
# 4 . Find out in this data that how many unique id's we have 

print("The unique ID's found are ", len(fitbit_dt.Id.unique()))
pd.DataFrame(fitbit_dt.Id.unique())
```

```python
# 5 . which id is one of the active id that you have in whole dataset 

# Taking the person who has more total steps as active person

fg = pd.DataFrame(fitbit_dt.groupby('Id')['TotalSteps'].mean())
fh = fg.sort_values('TotalSteps', ascending = False)
fh.head(1)
```

```python
# 6 . how many of them have not logged there activity find out in terms of number of ids 
k1 = pd.DataFrame(fitbit_dt.groupby(['Id'])['LoggedActivitiesDistance'].sum())
k2 = k1[k1['LoggedActivitiesDistance'] == 0]
print('The number of ids which have not logged there activity distance is', len(k2))
k2
```

```python
# 8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset 

# Assuming on an average 2000 calorie burn is required per day for a man/woman, number of fit persons are as folloeing :

gt = fitbit_dt.groupby(['Id'])[['Calories']].mean()
k = gt[gt['Calories'] > 2000]
print('Number of Fit people in our dataset are', len(k))
k
```

```python
# 9. how many person are not a regular person with respect to activity try to find out those

za = pd.DataFrame(fitbit_dt['ActivityDate'].unique())
print('Total number of unique days in dataset are', len(za))
#out of 32 days, if person is absent for more than 14 days, it will be considered unregular.

gt = fitbit_dt.groupby('Id')[['ActivityDate']].count()
st = gt[gt['ActivityDate']<14]
print('Number of people not regular are', len(st))
st
```

```python
# 10 . who is the third most active person in this dataset find out those in pandas and in sql both .  
fg = pd.DataFrame(fitbit_dt.groupby('Id')['TotalSteps'].mean())
fh = fg.sort_values('TotalSteps', ascending = False)
zs = fh.head(3)
#zs.iloc[[2]]
zs[2:3]
```
