```python
agent_perf['Date1'] = pd.to_datetime(agent_perf['Date'])
```

```python
df1 = agent_perf.sort_values(by='Date1', ascending=True)
```

```python
##1. Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave
agent_perf[['Agent Name','Average Rating']].groupby('Agent Name').sum()/5
```

```python
agent_perf.groupby(['Agent Name','week'])['Average Rating'].mean()
```

```python
##2. Total working days for each agents 
agent_perf.groupby(['Agent Name'])[['day']].count()
```

```python
#4. total Feedback that you have received
agent_perf.groupby(['Agent Name'])[['Total Feedback']].sum()
```

```python
##8. how many feedaback agents have received more then 4.5 average
df3 = agent_perf.groupby(['Agent Name'])['Total Feedback'].mean()>4.5
dff3=pd.DataFrame(df3)
dff3[dff3["Total Feedback"] == True].value_counts()
```

```python
# 9 . average weekly response time for each agent 
agent_perf['Average_response_time_sec'] = agent_perf['Average Response Time'].str.split(':').apply(lambda x: int(x[1]) * 60 + int(x[2]))
agent_perf.groupby(['Agent Name', 'week'])['Average_response_time_sec'].mean()
```

```python
agent_perf['Avg response time'] = pd.to_datetime(agent_perf['Average Response Time'])
```

```python
#agent_perf['Av_res_tm'] = pd.to_datetime(agent_perf['Average Response Time'],format='%H:%M:%S').dt.minute
agent_perf.groupby(['Agent Name','week'])["Avg response time"].mean()
```

```python
##10. average weekely resolution time for each agents
agent_perf['Average_resolution_time_sec'] = agent_perf['Average Resolution Time'].str.split(':').apply(lambda x: int(x[1]) * 60 + int(x[2]))
agent_perf.groupby(['Agent Name', 'week'])['Average_resolution_time_sec'].mean()
```
