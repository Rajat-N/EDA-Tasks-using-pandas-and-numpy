```python
# 11 . Group by region and find out which region is more profitable 
ok = pd.merge(orders,users, how='left', on='Region')
oi=pd.DataFrame(ok.groupby(['Region'])[['Region','Profit']].sum())
oi
```

```python
# 12 . Try to find out overall in which country we are giving more discount

ko = pd.DataFrame(orders.groupby(['State or Province'])['Discount'].sum())
ki = ko.sort_values(by='Discount', ascending=False)
ki.head(3)
```
